## Constrained Rosenbrock optimization
This code is a optimization of Constrained Rosenbrock function

```python
import jax
import jax.numpy as jnp
import numpy as np
import scipy.optimize

# 1. Enable 64-bit for JAX to match SciPy's float64 expectation
jax.config.update("jax_enable_x64", True)

# Rosenbrock function
def rosenbrock(x, y):
    a = 1
    b = 0
    return a * (y - x**2)**2 + (b - x)**2

# Loss function using jnp
rosenbrock_loss = lambda x: rosenbrock(x[0], x[1])

# Grad function - we wrap it in np.array to satisfy SLSQP
rosen_grad = lambda x: np.array(jax.grad(rosenbrock_loss)(x))

# Constraint value
c = 1.0

# Equality constraint:
# Using jax.jacobian for the constraint makes it even more robust
cons_fun = lambda x: x[0]**2 + x[1]**2 - c
cons_jac = lambda x: np.array(jax.jacobian(cons_fun)(x))

constraint = {
    'type': 'eq',
    'fun': cons_fun,
    'jac': cons_jac  # Adding the JAX-computed Jacobian here
}

# Initial guess
np.random.seed(0)
init_guess = np.random.randn(2)
x0 = np.array(init_guess) # Keep the initial guess as a standard numpy array

result = scipy.optimize.minimize(
    rosenbrock_loss, 
    x0, 
    method='SLSQP', 
    jac=rosen_grad,
    constraints=constraint
)

print(result)
```
